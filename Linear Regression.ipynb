# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
#reading the data

dataset = pd.read_csv("Data Set\data.csv")
```

```python
pd.set_option("display.max_columns", None)
dataset.head()
```

```python
size = dataset['LotArea']
price= dataset['SalePrice']
```

```python
#converting the dataframes into arrays
x = np.array(size).reshape(-1,1)
y = np.array(price).reshape(-1,1)
```

```python
x
```

```python
y
```

```python
#fitting into the model
model = LinearRegression()
model.fit(x,y)
```

```python
#Checking the error values(MSE,R)

mse = mean_squared_error(x,y)
print("The Mean Squared Error = ",math.sqrt(mse))
print("R value is = ",model.score(x,y))
```

```python
print(model.coef_[0])
```

```python
print(model.intercept_[0])
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,model.predict(x),color="yellow")
plt.title("Linear Regression")
plt.xlabel("Size")
plt.ylabel("Price")
```

```python
print("Prediction by the model for LotArea 14260 = ",model.predict([[14260]]))
```
